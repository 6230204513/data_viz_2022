## ต่อตาราง

* ต่อในแนวแกน Y (เพิ่มจำนวนข้อมูล)
* ต่อในแนวแกน X (เพิ่มรายละเอียดของข้อมูล)

```python
BKK = data_df[data_df['province_of_onset']=='กรุงเทพมหานคร'] #แสดงข้อมูลที่มี province_of_onset เป็น กรุงเทพมหานคร
BKK.head()
```

```python
KKC = data_df[data_df['province_of_onset']=='ขอนแก่น'] #แสดงข้อมูลที่มี province_of_onset เป็น ขอนแก่น
KKC.head()
```

## ต่อตารางให้เป็นข้อมูล 2 จังหวัด

```python
dataBKKKKC = pd.concat([BKK,KKC]) #นำตารางข้อมูลที่มี province_of_onset เป็น กรุงเทพมหานคร และ ขอนแก่น มาต่อกัน
dataBKKKKC #แสดงตารางข้อมูล
```

### ต่อแกน X

```python
personal_data = data_df[['No.', 'sex', 'age', 'nationality']]
personal_data.head()
```

```python
covid_data = data_df [['No.', 'announce_date', 'province_of_isolation', 'province_of_onset', 'risk']]
covid_data.head()
```

```python
covid_data = data_df [['No.', 'announce_date', 'province_of_isolation', 'province_of_onset', 'risk']].sample(frac=1)  # สลับ row สุ่มมั่ว โดยใช้ .sample(frac=1)
covid_data.head()
```

### เอาตารางสองอันมาต่อกัน

```python
personalcovid_data = personal_data.merge(covid_data) # .merge เป็นคำสั่งที่ใช้ในการต่อ row เมื่อ หัวข้อตารางแตกต่างกัน ก็ยังสามารถนำมาต่อกันได้ อีกทั้งยังเพิ่มหัวข้อตารางตามที่ดึงข้อมูลมาด้วย
personalcovid_data.head()
```

```python
personalcovid_data = covid_data.merge(personal_data)
personalcovid_data.head()
```
