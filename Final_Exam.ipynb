# เตรียมข้อมูล

```python
import pandas as pd # import ฟังก์ชัน pandas
import os # จัดการเกี่ยวกับไฟล์ต่างๆ ในโฟล์เดอร์
```

```python
from google.colab import drive # เชื่อม drive ของเรา ถ้าเชื่อมสำเร็จจะขึ้นคำว่าMounted at /content/drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/My Drive/ovrdt_data' # ชี้ว่า data เราอยู่โฟล์เดอร์ไหน
```

```python
data_file_path = os.path.join(path,'alldb.csv') # เอาที่อยู่ของไฟล์เชื่อมกับที่อยู่ของไฟล์เรา โดยกำหนดชื่่อไฟล์ที่เราต้องการ
data_alldb = pd.read_csv(data_file_path)
```

```python
data_alldb
```

```python
data_file_path2 = os.path.join(path,'hospital_list.xlsx')
data_hospital = pd.read_excel(data_file_path2)
```

```python
data_hospital
```

# จัดการข้อมูล

```python
new_alldb = data_alldb[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey']] # ตัดให้เหลือ column ที่ต้องการนำไปใช้
new_alldb
```

```python
new_alldb.isnull().sum() # เช็คค่า missing 
```

```python
drop_alldb = new_alldb.dropna() # จัดการค่า missing โดยการ drop ทิ้งไป
drop_alldb
```

```python
# เช็คข้อมูลหลังจากการ drop ว่าหายไปกี่เปอร์เซ็นต์
removes = new_alldb.shape[0] - drop_alldb.shape[0]
print(f'size before drop = {new_alldb.shape[0]}')
print(f'size after drop = {drop_alldb.shape[0]}')
print(f'% from drop = {(removes/new_alldb.shape[0])*100}')
```

## จัดการให้ค่า site เป็นตัวเลข 5 หลัก(ตัวเลขเท่านั้น)

```python
set(drop_alldb['site']) # set ดูค่าที่เก็บไว้
```

```python
drop_alldb = drop_alldb[(drop_alldb['site'].astype(str).str.isdigit()) & (drop_alldb['site'].str.len() == 5)] # เช็ค type ของข้อมูลโดยใช้ str.isdigit() และทำการเลือกข้อมูลที่เป็นตัวเลข 5 หลักเท่านั้น
drop_alldb
```

```python
new_hospital = data_hospital[['hcode','name','zone_code','zone_name','provincecode','province','amphurcode','amphur','tamboncode','tambon']] # ตัดให้เหลือ column ที่ต้องการนำไปใช้
new_hospital
```

```python
new_hospital.isnull().sum() # เช็คค่า missing
```

```python
drop_hospital = new_hospital.dropna() # จัดการค่า missing โดยการ drop 
drop_hospital
```

```python
# เช็คข้อมูลหลังจากการ drop ว่าหายไปกี่เปอร์เซ็นต์
removes = new_hospital.shape[0] - drop_hospital.shape[0]
print(f'size before drop = {new_hospital.shape[0]}')
print(f'size after drop = {drop_hospital.shape[0]}')
print(f'% from drop = {(removes/new_hospital.shape[0])*100}')
```

```python
set(drop_hospital['hcode']) # set ดูค่าที่เก็บไว้
```

## รวมตารางข้อมูล alldb และ hospital เข้าด้วยกัน โดยใช้การ merge

```python
data = drop_alldb.merge(drop_hospital,how='left',left_on='site',right_on='hcode') # เชื่อมตารางโดยตารางทางซ้ายเป็น alldb มี key คือ site เชื่อมกับ hospital มี key คือ hcode
data
```

```python
data.isnull().sum() # เช็คค่า missing
```

```python
drop_data = data.dropna() # จัดการค่า missing โดยการ drop
drop_data
```

```python
# เช็คข้อมูลหลังจากการ drop ว่าหายไปกี่เปอร์เซ็นต์
removes = data.shape[0] - drop_data.shape[0]
print(f'size before drop = {data.shape[0]}')
print(f'size after drop = {drop_data.shape[0]}')
print(f'% from drop = {(removes/data.shape[0])*100}')
```

## เปลี่ยน type ข้อมูล

```python
drop_data.info() # เช็ค type ข้อมูล
```

```python
# ทำการเปลี่ย type ของข้อมูล
drop_data['RowKey'] = drop_data['RowKey'].astype('string')
drop_data['phone'] = drop_data['phone'].astype('string')
drop_data['urine'] = drop_data['urine'].astype('string')
```

## จัดการให้ค่า user เป็นตัวเลข 5 หลัก(ตัวเลขเท่านั้น)

```python
set(drop_data['user']) # set ดูค่าที่เก็บไว้
```

```python
drop_data = drop_data[(drop_data['user'].astype(str).str.isdigit()) & (drop_data['user'].str.len() == 5)] # เช็ค type ของข้อมูลโดยใช้ str.isdigit() และทำการเลือกข้อมูลที่เป็นตัวเลข 5 หลักเท่านั้น
drop_data
```

## เปลี่ยน date เป็น YYYY-MM-DD

```python
drop_data['date'] = pd.to_datetime(drop_data['date'],format="%Y/%m/%d") # เปลี่ยน format ของข้อมูลวันที่ให้อยู่ในรูปแบบเดียวกัน
drop_data[['date']]
```

## เปลี่ยน time เป็น hh-mm-ss

```python
drop_data['time'] = pd.to_datetime(drop_data['time'],format='%H:%M%:%S',errors='ignore') # เปลี่ยน format ของข้อมูลเวลาให้อยู่ในรูปแบบเดียวกัน
drop_data[['time']]
```

## จัดการให้ค่า ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
set(drop_data['ov']) # set ดูค่าที่เก็บไว้
```

```python
drop_data[drop_data['ov'] == 'ติดเชื้อ +2 '] # ดูว่า 'ติดเชื้อ +2 ' เป็นค่าที่เท่าไร
```

```python
drop_data['ov'] = drop_data['ov'].drop(16413) # drop ค่า 'ติดเชื้อ +2 ' ทิ้ง
```

```python
drop_data = drop_data.dropna(subset=['ov']) # จัดการค่า missing 
```

```python
set(drop_data['ov']) # เช็คค่าที่เก็บไว้
```

## จัดการให้ค่า PartitionKey เป็นตัวเลข 5 หลัก(ตัวเลขเท่านั้น)

```python
set(drop_data['PartitionKey']) # set ดูค่าที่เก็บไว้
```

```python
drop_data = drop_data[(drop_data['PartitionKey'].astype(str).str.isdigit()) & (drop_data['PartitionKey'].str.len() == 5)] # เช็ค type ของข้อมูลโดยใช้ str.isdigit() และทำการเลือกข้อมูลที่เป็นตัวเลข 5 หลักเท่านั้น
drop_data
```
