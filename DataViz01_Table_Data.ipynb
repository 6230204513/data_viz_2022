```python
import os  #การจัดการเกี่ยวกับไฟล์ต่าง ๆ มันอยู่โฟลเดอร์ไหนให้ list ออกมา
```

```python
from google.colab import drive #เชื่อมต่อ google drive กับ Colab ของเรา นี้เป็นคำสั่งเชื่อม |accout colab กับ Drive ควรเป็น accout เดียวกัน|
drive.mount('/content/drive')
path = '/content/drive/My Drive/dataviz_2021_data' #มันอยู่ใน Driveของเรา บอกทางผ่านไฟล์เรา
```

```python
data_file_path = os.path.join(path,'pm-20-jan-2021.csv')  # ความสำคัญ เราทำบน window linux มันจะไม่เหมือนกัน มันเอาไฟล์มา/กัน กลับมาเขียนต่อ
data_df = pd.read_csv(data_file_path)
```

```python
data_df
```

```python
data_df['risk'].unique() #ตรวจสอบดูว่า risk ประกอบไปด้วยอะไรบ้าง
```

```python
data_df.isnull().any() #ตรวจสอบค่า Missing
```

```python
data_new = data_df[['age']].mean() #แทนที่ค่า missing ของ age ด้วยค่า mean
data_df['age'] = data_df[['age']].fillna(data_new)
data_df.isnull().any() #ตรวจสอบ missing อีกครั้ง
```

```python
data_df['sex'].unique() #ตรวจสอบค่า sex ทั้งหมดว่าประกอบด้วยอะไรบ้าง
```

```python
nan_sex = data_df[['sex']].mode() #แทนที่ค่า missing ของ sex ด้วยค่า mode
data_df['sex'] = data_df[['sex']].fillna(nan_sex)
data_df.isnull().any() #ตรวจสอบ missing อีกครั้ง
```

```python
RA = data_df[data_df['risk']=='สนามมวย'] #กำหนดให้ RA คือ risk ที่เป็น สนามมวย
RA #แสดงผล
```

```python
RB = data_df[data_df['risk']=='สถานบันเทิง'] #กำหนดให้ RB คือ risk ที่เป็น สถานบันเทิง
RB #แสดงผล
```

```python
RC = data_df[data_df['risk']=='สัมผัสผู้ป่วยยืนยัน'] #กำหนดให้ RC คือ risk ที่เป็น สัมผัสผู้ป่วยยืนยัน
RC #แสดงผล
```

```python
R_ABC = pd.concat([RA,RB,RC]) #นำ case ของสนามมวย สถานบันเทิงและสัมผัสผู้ป่วยยืนยัน มารวมกันเป็นตาราง R_ABC
R_ABC #แสดงผล
```

```python
table_risk = R_ABC[['sex', 'age', 'risk']] #สร้างตาราง table_risk ที่ประกอบไปด้วย 'sex', 'age', 'risk'
table_risk #แสดงผล
```

```python
#แทนที่ชายด้วยเลข 0 หญิงแทนด้วยเลข 99 กำหนดให้เป็นค่าที่ห่างกันเนื่องจากจะทำให้กราฟแสดงผลดูง่ายขึ้น
table_risk['sex'].replace({'ชาย':0,'หญิง':99}, inplace=True) # replace คือคำสั่งแทนที่ค่า inplace=True คือการเปลี่ยน ชาย ให้เป็น 0 และเปลี่ยนหญิงให้เป็น 99
```

```python
table_risk['sex'].unique() #แสดงค่าใน sex ว่าประกอบไปด้วยอะไรบ้าง
```

```python
#แทนที่ค่าใน column risk ด้วยคำสั่ง replace โดยให้ inplace=True เพื่อเเทนที่ 'สนามมวย'ด้วย 'boxing stadium'/'สถานบันเทิง'ด้วย 'entertainment place'/ 'สัมผัสผู้ป่วยยืนยัน'ด้วย 'touch the patient(confirmed result) 
table_risk['risk'].replace({'สนามมวย': 'boxing stadium','สถานบันเทิง': 'entertainment place', 'สัมผัสผู้ป่วยยืนยัน': 'touch the patient(confirmed result)'}, inplace=True) #เนื่องจากกราฟไม่แสดงชื่อภาษาไทย ดังนั้นจึงต้องตั้งชื่อเป็นภาษาอังกฤษ
```

```python
table_risk['risk'].unique() #แสดงค่าใน risk ว่าประกอบไปด้วยอะไรบ้าง
```

```python
pd.plotting.parallel_coordinates(table_risk,'risk', color = ['#FF00E0', '#00FFE2', '#ECFF00']) #กำหนดสีเอง ให้กับแต่ละ ค่าของ risk
```
