### 1. เขียน function ที่รับ input เป็น output ของ boxplot แล้ว แสดงค่า min, max, Q1, Q2, Q3


```python
ybp = plt.boxplot(df['SepalLength'], meanline = True ,showmeans=True)
```

```python
#เขียน function ที่รับ input เป็น output ของ boxplot แล้ว แสดงค่า min, max, Q1, Q2,Q3
def box_vals(ybp) :
        q1 = ybp['boxes'][0].get_xdata()[0]
        q2 = ybp['medians'][0].get_xdata()[0]
        q3 = ybp['boxes'][0].get_xdata()[2]
        mmin = min(ybp['caps'][0].get_xdata())
        mmax = max(ybp['caps'][1].get_xdata())
        print(f' ค่าของ Q1 เท่ากับ {q1} \n ค่าของ Q2 เท่ากับ {q2} \n ค่าของ Q3 เท่ากับ{q3} \n ค่า min เท่ากับ {mmin} \n ค่า max เท่ากับ {mmax}')
```

```python
box_vals(ybp)
```
